## Preprocessing Text

#### Creating a function to convert the text in lowercase, remove the extra space, special chr., ulr and links.

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+',b'',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

#### Defining dependent and independent variable as x and y

```python
x = data['text']
y = data['class']
```

## Training the model

#### Splitting the dataset into training set and testing set.

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

## Testing the Model

```python
def output_lable(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)

    return print("\n\nLR Predicition: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction:{}".format(output_lable(pred_LR[0]),
                                                                                                             output_lable(pred_DT[0]),
                                                                                                             output_lable(pred_GB[0]),
                                                                                                             output_lable(pred_RF[0])))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("Loading datasets...")
true_df = pd.read_csv('True.csv', engine='c', lineterminator='\n', on_bad_lines='skip')
fake_df = pd.read_csv('Fake.csv', engine='c', lineterminator='\n', on_bad_lines='skip')

# Standardizing column names to lowercase and remove hidden spaces
true_df.columns = true_df.columns.str.strip().str.lower()
fake_df.columns = fake_df.columns.str.strip().str.lower()

print(f"Loaded {len(true_df)} Real articles and {len(fake_df)} Fake articles.")

# Add labels
true_df['Article_Type'] = 'Real'
fake_df['Article_Type'] = 'Fake'

# Combine
combined_df = pd.concat([true_df, fake_df], ignore_index=True)


if 'date' not in combined_df.columns:
    print("CRITICAL ERROR: 'date' column missing")
    print(combined_df.columns.tolist())
else:
    print("Parsing dates ...")
    combined_df['date'] = pd.to_datetime(combined_df['date'], errors='coerce')
    combined_df = combined_df.dropna(subset=['date'])

    print(f"Data remaining after date cleaning: {len(combined_df)} articles.")

    # Extract Year and Month
    combined_df['YearMonth'] = combined_df['date'].dt.to_period('M')

    # Group by month and count
    print("Generating trend data...")
    trend_data = combined_df.groupby(['YearMonth', 'Article_Type']).size().unstack(fill_value=0)


    if 'Fake' not in trend_data.columns:
        trend_data['Fake'] = 0
    if 'Real' not in trend_data.columns:
        trend_data['Real'] = 0

    # Plotting
    plt.figure(figsize=(14, 7))
    plt.plot(trend_data.index.astype(str), trend_data['Fake'], color='#ef4444', marker='o', linewidth=2, label='Fake News')
    plt.plot(trend_data.index.astype(str), trend_data['Real'], color='#22c55e', marker='s', linewidth=2, label='Real News')

    plt.title('Trend Analysis: Fake vs. Real News Articles Over Time', fontsize=16, fontweight='bold')
    plt.xlabel('Timeline (Year-Month)', fontsize=12)
    plt.ylabel('Number of Articles Published', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()

    plt.show()
```
